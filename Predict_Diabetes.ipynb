```python
import tensorflow as tf
import pandas as pd
```

```python
file_path = 'C:\\Users\\User\\Desktop\\Diabetes_updated_data.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
#Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data_cleaned = data.dropna()

# Check if missing values still exist
missing_values_after = data_cleaned.isnull().sum()
print("Missing values after cleaning:")
print(missing_values_after)
```

```python
data_cleaned.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = ['gender', 'smoking_history', 'location']

# Use .loc[] to avoid the warning and properly modify the DataFrame
for column in categorical_columns:
    data_cleaned.loc[:, column] = label_encoder.fit_transform(data_cleaned[column])

# Check the first few rows of the cleaned data
data_cleaned.head()

```

```python
data = data_cleaned.drop(['year'], axis=1)
data.head()
```

```python
# Check correlation between year and diabetes (categorical)
print(data.corr())

# Check if there are patterns in location affecting diabetes
print(data.groupby('location')['diabetes'].mean())
```

```python
#Split the dataset
from sklearn.model_selection import train_test_split

X = data.drop('diabetes', axis=1)
Y = data['diabetes']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

```python
#Train the model
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(random_state = 42)

classifier.fit(X_train, Y_train)
```

```python
Y_prediction = classifier.predict(X_test)

print("Predicted values:")
print(Y_prediction)
```

```python
import numpy as np

# Example new data (gender: Female, age: 45, etc.)
new_data = {
    'gender': 'Female',
    'age': 45,
    'location': 0,
    'race:AfricanAmerican': 0,
    'race:Asian': 0,
    'race:Caucasian': 1,
    'race:Hispanic': 0,
    'race:Other': 0,
    'hypertension': 1,
    'heart_disease': 0,
    'smoking_history': 'never',
    'bmi': 25.3,
    'hbA1c_level': 5.7,
    'blood_glucose_level': 130
}

# Convert new_data into a DataFrame
new_data_df = pd.DataFrame([new_data])

# Encode the 'gender' and 'smoking_history' columns with the same LabelEncoder used earlier
new_data_df['gender'] = label_encoder.fit_transform(new_data_df['gender'])
new_data_df['smoking_history'] = label_encoder.fit_transform(new_data_df['smoking_history'])

# Ensure the order of columns matches the training data
new_data_df = new_data_df[X.columns]

# Step 3: Predict diabetes for the new data
new_prediction = classifier.predict(new_data_df)

# Display the result
print(f"Predicted diabetes (0 = No, 1 = Yes): {new_prediction[0]}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_prediction)
print(f"Accuracy: {accuracy:.2f}")

# Confusion matrix
confusion_matrix = confusion_matrix(Y_test, Y_prediction)
print("Confusion Matrix:")
print(confusion_matrix)

# Classification report
class_report = classification_report(Y_test, Y_prediction)
print("Classification Report:")
print(class_report)
```
